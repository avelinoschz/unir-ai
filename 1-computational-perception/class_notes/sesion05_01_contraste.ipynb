```python
import cv2 as cv

img01=cv.imread('../imagenes/flowers.jpeg')
img02=cv.imread('../imagenes/flowers.jpeg',cv.IMREAD_GRAYSCALE)
```

```python
from matplotlib import pyplot as plt
plt.imshow(cv.cvtColor(img01, cv.COLOR_BGR2RGB))
plt.title('imagen')
plt.show()
```

```python
plt.imshow(cv.cvtColor(img02, cv.COLOR_BGR2RGB))
plt.title('imagen')
plt.show()
```

```python
import numpy as np
M = np.ones(img02.shape, dtype = "uint8") * 30 
```

```python
M
```

## Operaciones bsica entre imgenes

```python

img_suma = np.clip(img02.astype(np.int16) + M, 0, 255).astype(np.uint8)
```

```python
plt.imshow(cv.cvtColor(img_suma, cv.COLOR_BGR2RGB))
plt.title('imagen')
plt.show()
```

```python
img_resta = np.clip(img02.astype(np.int16) - M, 0, 255).astype(np.uint8)
```

```python
plt.imshow(cv.cvtColor(img_resta, cv.COLOR_BGR2RGB))
plt.title('imagen')
plt.show()
```

```python
constante_M = np.ones_like(img01) * 100

img_suma_color = np.clip(img01.astype(np.int16) + constante_M, 0, 255).astype(np.uint8)
```

```python
plt.imshow(cv.cvtColor(img_suma_color, cv.COLOR_BGR2RGB))
plt.title('imagen')
plt.show()
```

```python

# Definir constantes a sumar a cada canal
c_azul = 20
c_verde = 30
c_rojo = 100

canal_azul, canal_verde, canal_rojo = cv.split(img01)

c_azul_const = np.clip(canal_azul.astype(int) + c_azul, 0, 255).astype(np.uint8)
c_verde_const = np.clip(canal_verde.astype(int) + c_verde, 0, 255).astype(np.uint8)
c_rojo_const = np.clip(canal_rojo.astype(int) + c_rojo, 0, 255).astype(np.uint8)

image_final = cv.merge((c_azul_const, c_verde_const, c_rojo_const))

plt.imshow(cv.cvtColor(image_final, cv.COLOR_BGR2RGB))
plt.title('imagen')
plt.show()
```
