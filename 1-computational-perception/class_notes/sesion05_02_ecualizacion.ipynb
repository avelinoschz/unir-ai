```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Cargar la imagen
image = cv2.imread('../imagenes/ninia con perro.jpeg')

# Dividir la imagen en los canales de color
b, g, r = cv2.split(image)

# Ecualizar cada canal de color por separado
equ_b = cv2.equalizeHist(b)
equ_g = cv2.equalizeHist(g)
equ_r = cv2.equalizeHist(r)

# Fusionar los canales nuevamente en la imagen ecualizada
equ_image = cv2.merge([equ_b, equ_g, equ_r])

# Mostrar la imagen original y la imagen ecualizada
cv2.imshow('Original', image)
cv2.imshow('Ecualizada', equ_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(equ_image, cv2.COLOR_BGR2RGB))
```
